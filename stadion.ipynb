# Analiza stadionov s kapaciteto večjo od 40 000


Za projekt pri predmetu Programiranje1 bom analizirala podatke največjih stadionov. Podatke sem dobila iz spletne strani `https://en.wikipedia.org/wiki/List_of_stadiums_by_capacity`. Obseg zajema stadione s kapaciteto večjo od 40 000.

Za vsak stadion sem pridobila naslednje podatke:
    - ime stadiona
    - kapaciteta stadiona
    - mesto v katerem je stadion
    - država iz katere je ta stadion
    - ekipa, ki trenira na tem stadionu
    - kateri športi se lahko izvajajo na tem stadionu
    - kontinent

```python
# potrebni paketi za obdelavo podatkov
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# naložimo razpredelnico, iz katere pridobivamo podatke
stadioni = pd.read_csv('stadioni.csv')
```

```python
#razpredelnica stadionov
stadioni.head(10)
```

```python
# zadnji del razpredelnice
stadioni.tail()
```

```python
# ime in kapaciteta vsakega stadiona
stadioni[['ime', 'kapaciteta']]
```

```python
# podatki največjega stadiona
stadioni.iloc[0]
```

```python
# podatki 108-ega po vrsti stadiona
stadioni.loc[100]
```

```python
stadioni[501 : 510].kapaciteta == 40000
```

```python
# razpredelnica stadionov s kapaciteto 40 000 
stadioni[stadioni.kapaciteta == 40000]
```

```python
# Razpredelnica desetih stadionov, na katerih je možno igrati več različnih športov
stadioni.sort_values([
    'uporaba3', 'uporaba2', 'uporaba1'
], ascending=True)[
    ['ime', 'kapaciteta', 'uporaba1', 'uporaba2', 'uporaba3']
].head(10)
```

```python
# glava razpredelnice pogruperane po državah
stadioni.groupby('drzava').count().head()
```

```python
# nov stoplec z zaokroženimi kapacitetami na 500
stadioni['zaokrozena kapaciteta'] = 1500 * (stadioni['kapaciteta'] // 1500)
```

```python
stadioni
```

```python
zaokrozena_kapaciteta = stadioni.groupby('zaokrozena kapaciteta').count()
```

```python
stadioni.groupby('zaokrozena kapaciteta').count()
```

Opazimo, da je največ stadionov s približno kapaciteto od 51 000 do 39 000

```python
# razpredelnica z naraščajočo zaokroženo kapaciteto po državah
zaokrozena_kapaciteta.sort_values(['drzava'], ascending=False)
```

```python
stadioni.groupby('kontinent')[
    'kapaciteta'
].size().plot.bar(x='kontinent', y='kapaciteta')
```

Opazimo, da je v Severni Ameriki največ stadionov s kapaciteto večjo od 40 000.
Zato bom sedaj opazovala malo bolj stadione Severne Amerike in potem se Azije.

```python
stadioni[
    stadioni['kontinent'] == 'North America'
].groupby('zaokrozena kapaciteta').size().plot.bar(x='zaokrozena kapaciteta', figsize=(20, 5))
```

Iz grafa opazimo, da je v Severni Ameriki stevilo stadionov z zaokroženo kapaciteto na 40 500 daleč največ, ostalih pa je precej manj.

```python
stadioni[
    stadioni['kontinent'] == 'Asia'
].groupby('zaokrozena kapaciteta').size().plot.bar(x='zaokrozena kapaciteta', figsize=(20, 5))
```

Iz grafa opazimo, da je v Aziji večino stadionov s približno kapaciteto 39 000, 45 000, 49 500 in 60 000. Ostalih pa je precej manj. 

```python
# Stoplični diagram za stadione s kapaciteto 40 000
stadioni[
    stadioni['kapaciteta'] == 40000
].groupby('kontinent')[
    'zaokrozena kapaciteta'
].size().plot.bar(x='kontinent', y='zaokrozena kapaciteta')
```

```python
stadioni[stadioni['kapaciteta'] == 40000].groupby('kontinent').size().plot(kind='pie', figsize=(5, 5), fontsize=12)
```

Opazimo, da je v Aziji največ stadionov s kapaciteto enako 40 000 in v Evropi nimamo prav nobenega s tako kapaciteto.

```python
stadioni[stadioni['kapaciteta'] >= 80000].groupby('kontinent').size().plot(kind='pie', figsize=(5, 5), fontsize=12)
```

```python
stadioni[stadioni['kapaciteta'] >= 80000].groupby('kontinent').count()
```

Opazimo, da je daleč največ stadionov, s kapaciteto večjo ali enako 80 000, v Severni Ameriki kar 17 več kot na naslednjem kontinentu, medtem ko je največji stadion v Aziji.
